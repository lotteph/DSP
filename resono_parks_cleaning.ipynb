```python
import glob
import numpy as np
import pandas as pd
from datetime import timedelta
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='once')
```

```python
df = pd.read_csv('resono_parks.csv')
```

```python
df1 = df.drop(['Unnamed: 0', 'Locationcode', 'End_Dates', 'End_Time', 'End'], axis=1)
df1.drop_duplicates(inplace=True)
df1.reset_index(inplace=True, drop=True)
df1['Start'] = pd.to_datetime(df1['Start'], utc=True)
df1['Start'] = pd.to_datetime(df1['Start'].dt.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
df1['End'] = df1['Start'] + dt.timedelta(minutes=15, seconds=0)
df1.drop('Start', axis=1, inplace=True)
```

```python
df1 = df1.rename({'Start_Dates': 'End_Dates', 'Start_Time' : 'End_Time'},
                                                   axis=1) 
df_average_visits = df1.copy() 
df_average_visits = df_average_visits.groupby(['Location', 'End']).mean().reset_index()
```

```python
df_average_visits['Visits'] = df_average_visits["Visits"].astype(int)
```

```python
resono_data = df_average_visits.copy()
```

```python
resono_data['End_Dates'] = pd.to_datetime(resono_data['End']).dt.date
resono_data['End_Time'] = pd.to_datetime(resono_data['End']).dt.time
```

```python

resono_data = resono_data[resono_data.Location != 'Frankendael']
resono_data.groupby('Location').count()
```

```python
resono_data.to_csv('test.csv')
```
