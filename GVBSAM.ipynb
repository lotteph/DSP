# Journey Destination Date

```python
# Drop NaN values and set date to datetime type
Journey_Destination_Date_Cleaned = Journey_Destination_Date.dropna()
Journey_Destination_Date_Cleaned['Datum'] = pd.to_datetime(Journey_Destination_Date_Cleaned['Datum'])

# Get latitude & longitude per station
LatLong_Per_Station = Journey_Destination_Date_Cleaned.drop_duplicates(subset=['AankomstHalteNaam']).set_index('AankomstHalteNaam')[['AankomstLat', 'AankomstLon']]

Journey_Destination_Date_Cleaned
# Get Journey per station
Journeys_Per_Station_Per_Day = Journey_Destination_Date_Cleaned.groupby(['Datum', 'AankomstHalteNaam']).agg({'AankomstLat':'mean', 'AankomstLon':'mean','AantalReizen':'sum'})

# Get all travels from 2020-09-21 and sort 
Journeys_2020_09_21 = Journeys_Per_Station_Per_Day.loc['2020-09-21']

# Sort by most popular stations
Journeys_2020_09_21.sort_values(by=['AantalReizen'], ascending=False, inplace=True)

# Drop date index and get top 20 most crowded stations
Journeys_2020_09_21_plot = Journeys_2020_09_21.droplevel('Datum').reset_index()[:10]

plt.figure(figsize=(40,10))
sns.barplot(x="AankomstHalteNaam", y="AantalReizen", data=Journeys_2020_09_21_plot);

# SUGGESTION: WITH LON + LAT, ONLY SELECT STATION 1KM OR LESS FROM PARK. 
```
