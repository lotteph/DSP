```python
import pandas as pd
import cbsodata

#If cbsodata is not installed --> Uncomment line below and run
#!pip install cbsodata

# Downloaden van tabeloverzicht
toc = pd.DataFrame(cbsodata.get_table_list())

# Downloaden van gehele tabel (kan een halve minuut duren)
data = pd.DataFrame(cbsodata.get_data('83765NED'))

# Downloaden van metadata
metadata = pd.DataFrame(cbsodata.get_meta('83765NED', 'DataProperties'))
```

```python
#Make a copy of the data and filter on Amsterdam
df = data.copy()
df_ams = df[df['Gemeentenaam_1'] == 'Amsterdam                               ']

#Select useful columns and reset index
cols = ['WijkenEnBuurten', 'SoortRegio_2',
       'Codering_3', 'IndelingswijzigingWijkenEnBuurten_4','AantalInwoners_5', 'Bevolkingsdichtheid_33','Mannen_6', 'Vrouwen_7', 'k_0Tot15Jaar_8',
       'k_15Tot25Jaar_9', 'k_25Tot45Jaar_10', 'k_45Tot65Jaar_11',
       'k_65JaarOfOuder_12']
df = df_ams[cols].reset_index(drop=True)
```

```python
#Sort by most inhabitants per neighbourhood in Amsterdam

ams_population_sorted = df.sort_values(by=['AantalInwoners_5', 'Bevolkingsdichtheid_33'], ascending=False)
ams_population_sorted.head(5)
```

```python
#Sort by highest populationdensity per neighbourhood in Amsterdam

ams_population_sorted = df.sort_values(by=['Bevolkingsdichtheid_33', 'AantalInwoners_5'], ascending=False)
ams_population_sorted.head(5)
```
