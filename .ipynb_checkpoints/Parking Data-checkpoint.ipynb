```python
#open specific json file
import os
import json
from math import radians, cos, sin, asin, sqrt
script_dir = os.path.dirname('parking-availability-f7b5e541-4e05-4c5f-901b-14b72ce9027c.json')
file_path = os.path.join(script_dir, 'Parking 2020-10/2020-10-18/18/parking-availability-f7b5e541-4e05-4c5f-901b-14b72ce9027c.json')
with open(file_path, 'r') as fi:
    data = json.load(fi)
    
```

```python
import pandas as pd
#make json file into pd dataframe
df = pd.json_normalize(data, record_path =['features'])
# drop irrelevant columns
df_selected = df.drop(['geometry.type', 'properties.State', 'properties.Type', 'type'], axis= 1)

df_selected['properties.FreeSpaceShort']=df_selected['properties.FreeSpaceShort'].apply(int)
df_selected['properties.ShortCapacity']=df_selected['properties.ShortCapacity'].apply(int)
df_selected['capacity.filled'] = (df_selected['properties.FreeSpaceShort']) / df_selected['properties.ShortCapacity']
df_selected
```

```python
Parking_coordinates = df_selected[['properties.Name', 'geometry.coordinates']].copy()
Parking_coordinates[['lat','lng']] = pd.DataFrame(Parking_coordinates['geometry.coordinates'].tolist(), index= Parking_coordinates.index)
Parking_coordinates = Parking_coordinates.set_index('properties.Name').rename_axis("Parking Location")[['lat','lng']]
Parking_coordinates.shape
```

```python
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r

def add_nearby_parkings(radius, center_point_dict, add_to_list):
    """
    radius in km
    center_point_dict needs to be a dictionary with the lon and lat from a location
    add_to_list specify to which list this needs to be added (for example vondelpark)
    """
    latpark = center_point_dict[0]['lat']
    lonpark = center_point_dict[0]['lng']

    # check for every station if it is within 1 km distance of the park
    for parking in range(len(Parking_coordinates)):
        name_parking = Parking_coordinates.iloc[parking].name
        lat_parking = Parking_coordinates.iloc[parking].lat
        lng_parking = Parking_coordinates.iloc[parking].lng

        a = haversine(lonpark, latpark, lng_parking, lat_parking)
        
        if a <= radius:
            add_to_list.append(name_parking)
```

```python
all_parks = ['vondelpark', 'oosterpark', 'sarphatipark',
               'westerpark', 'rembrandpark', 'erasmuspark']

vondelpark = [{'lat': 4.868648, 'lng': 52.357645}]
oosterpark = [{'lat': 4.920558, 'lng': 52.360098}]
sarphatipark = [{'lat': 4.896375, 'lng': 52.354364}]
westerpark = [{'lat': 4.873437, 'lng': 52.386986}]
rembrandpark = [{'lat': 4.846757, 'lng': 52.364145}]
erasmuspark = [{'lat': 4.851909, 'lng': 52.374808}]

# Create empty list that needs to be filled later
vondelpark_parkinglocations = []
oosterpark_parkinglocations = []
sarphatipark_parkinglocations = []
westerpark_parkinglocations = []
rembrandpark_parkinglocations = []
erasmuspark_parkinglocations = []

# calculate all station within 1 km from the park
add_nearby_parkings(2,vondelpark,vondelpark_parkinglocations)
add_nearby_parkings(2,oosterpark,oosterpark_parkinglocations)
add_nearby_parkings(2,sarphatipark,sarphatipark_parkinglocations)
add_nearby_parkings(2,westerpark,westerpark_parkinglocations)
add_nearby_parkings(2,rembrandpark,rembrandpark_parkinglocations)
add_nearby_parkings(2,erasmuspark,erasmuspark_parkinglocations)
```
