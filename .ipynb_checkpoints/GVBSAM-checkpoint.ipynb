# Exploration GBV Data 05/10/2020

```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import warnings

warnings.filterwarnings("ignore")
```

# Import CSV files

```python
Journey_Destination_Date = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Reis_Bestemming_Datum_20201005.csv", sep=";")
Journey_Destination_Hour = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Reis_Bestemming_Uur_20201005.csv", sep=";")
Journey_Departure_Date = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Reis_Herkomst_Datum_20201005.csv", sep=";")
Journey_Departure_Hour = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Reis_Herkomst_Uur_20201005.csv", sep=";")
Ride_Destination_Hour = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Rit_Bestemming_Uur_20201005.csv", sep=";")
Ride_Departure_Destination_Hour = pd.read_csv("GVB_DATA/2020/10/05/Datalab_Rit_Herkomst_Bestemming_Uur_20201005.csv", sep=";")
Ride_Departure_Hour = pd.read_csv("GVB_DATA/2020/10/05//Datalab_Rit_Herkomst_Uur_20201005.csv", sep=";")
```

# Journey Destination Date

```python
# Drop NaN values and set date to datetime type
Journey_Destination_Date_Cleaned = Journey_Destination_Date.dropna()
Journey_Destination_Date_Cleaned['Datum'] = pd.to_datetime(Journey_Destination_Date_Cleaned['Datum'])

# Get Journey per station
Journeys_Per_Station_Per_Day = Journey_Destination_Date_Cleaned.groupby(['Datum', 'AankomstHalteNaam']).agg({'AankomstLat':'mean', 'AankomstLon':'mean','AantalReizen':'sum'})

# Get all travels from 2020-09-21 and sort 
Journeys_2020_09_21 = Journeys_Per_Station_Per_Day.loc['2020-09-21']

# Sort by most popular stations
Journeys_2020_09_21.sort_values(by=['AantalReizen'], ascending=False, inplace=True)

# Drop date index and get top 20 most crowded stations
Journeys_2020_09_21_plot = Journeys_2020_09_21.droplevel('Datum').reset_index()[:10]

plt.figure(figsize=(40,10))
sns.barplot(x="AankomstHalteNaam", y="AantalReizen", data=Journeys_2020_09_21_plot);

# SUGGESTION: WITH LON + LAT, ONLY SELECT STATION 1KM OR LESS FROM PARK. 
```

# Journey Destination Hour

```python
# Drop NaN values and set date to datetime type and rename column
Journey_Destination_Hour_Cleaned = Journey_Destination_Hour.dropna()
Journey_Destination_Hour_Cleaned['Datum'] = pd.to_datetime(Journey_Destination_Hour_Cleaned['Datum'])
Journey_Destination_Hour_Cleaned.rename(columns={"UurgroepOmschrijving (van aankomst)": "AankomstTijd"}, inplace=True)

# Get Journey per station per hour
Journeys_Per_Station_Per_Hour = Journey_Destination_Hour_Cleaned.groupby(['Datum', 'AankomstHalteNaam', 'AankomstTijd']).agg({'AankomstLat':'mean', 'AankomstLon':'mean','AantalReizen':'sum'})

# Take 2020-09-21 data and station J.P. Heijestraat (very close to Vondelpark) as example
Journeys_Per_Station_Per_Hour_2020_09_21_plot = Journeys_Per_Station_Per_Hour.loc['2020-09-21', 'J.P. Heijestraat'].reset_index()

# Plot amount of journey arrivals per day
plt.figure(figsize=(40,10))
sns.lineplot(x="AankomstTijd", y="AantalReizen", data=Journeys_Per_Station_Per_Hour_2020_09_21_plot);
```

# Journey Departure Date

```python
# Drop NaN values and set date to datetime type
Journey_Destination_Departure_Date_Cleaned = Journey_Departure_Date.dropna()
Journey_Destination_Departure_Date_Cleaned['Datum'] = pd.to_datetime(Journey_Destination_Departure_Date_Cleaned['Datum'])

# Get journeys per day
Journeys_Per_Station_Departure_Per_Day = Journey_Destination_Departure_Date_Cleaned.groupby(['Datum', 'VertrekHalteNaam']).agg({'VertrekLat':'mean', 'VertrekLon':'mean','AantalReizen':'sum'})

# Get all departures from 2020-09-21 and sort 
Journeys_2020_09_21 = Journeys_Per_Station_Departure_Per_Day.loc['2020-09-21']
Journeys_2020_09_21.sort_values(by=['AantalReizen'], ascending=False, inplace=True)

# Drop date index and get top 20 most crowded stations
Journeys_2020_09_21_plot = Journeys_2020_09_21.droplevel('Datum').reset_index()[:10]

plt.figure(figsize=(40,10))
sns.barplot(x="VertrekHalteNaam", y="AantalReizen", data=Journeys_2020_09_21_plot);
```

# Journey Departure Hour

```python
# Drop NaN values and set date to datetime type and rename column
Journey_Departure_Hour_Cleaned = Journey_Departure_Hour.dropna()
Journey_Departure_Hour_Cleaned['Datum'] = pd.to_datetime(Journey_Departure_Hour_Cleaned['Datum'])
Journey_Departure_Hour_Cleaned.rename(columns={"UurgroepOmschrijving (van vertrek)": "VertrekTijd"}, inplace=True)

# Get Journey per station per hour
Journeys_Per_Station_Per_Hour = Journey_Departure_Hour_Cleaned.groupby(['Datum', 'VertrekHalteNaam', 'VertrekTijd']).agg({'VertrekLat':'mean', 'VertrekLon':'mean','AantalReizen':'sum'})

# Take 2020-09-21 data and station J.P. Heijestraat (very close to Vondelpark) as example
Journeys_Per_Station_Per_Hour_2020_09_21_plot = Journeys_Per_Station_Per_Hour.loc['2020-09-21', 'J.P. Heijestraat'].reset_index()

# Plot amount of journey arrivals per day
plt.figure(figsize=(40,10))
sns.lineplot(x="VertrekTijd", y="AantalReizen", data=Journeys_Per_Station_Per_Hour_2020_09_21_plot);
```

# Rides Desination Hour

```python
# Drop NaN values and set date to datetime type and rename column
Ride_Destination_Hour_Cleaned = Ride_Destination_Hour.dropna()
Ride_Destination_Hour_Cleaned['Datum'] = pd.to_datetime(Ride_Destination_Hour_Cleaned['Datum'])
Ride_Destination_Hour_Cleaned.rename(columns={"UurgroepOmschrijving (van aankomst)": "AankomstTijd"}, inplace=True)

# Get Journey per station per hour
Rides_Per_Station_Per_Hour = Ride_Destination_Hour_Cleaned.groupby(['Datum', 'AankomstHalteNaam', 'AankomstTijd']).agg({'AankomstLat':'mean', 'AankomstLon':'mean','AantalRitten':'sum'})

# Take 2020-09-21 data and station J.P. Heijestraat (very close to Vondelpark) as example
Rides_Per_Station_Per_Hour_2020_09_21_plot = Rides_Per_Station_Per_Hour.loc['2020-09-21', 'J.P. Heijestraat'].reset_index()

# Plot amount of journey arrivals per day
plt.figure(figsize=(40,10))
sns.lineplot(x="AankomstTijd", y="AantalRitten", data=Rides_Per_Station_Per_Hour_2020_09_21_plot);
```
