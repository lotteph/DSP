```python
import json
import glob
import numpy as np
import pandas as pd
from datetime import timedelta
import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = []
for name in glob.glob("KNMI (Weather) 2020/2020/0*/*/meteo-history*"):
    with open(name) as json_data:
        data = json.load(json_data)
        df = pd.DataFrame(data['waarneming'])
        dataframe.append(df) #append to the list
dataframe = pd.concat(dataframe, axis=0)
```

```python
# datum = date of the observation
# uur = hour of the observation
# windr = the direction of the wind (in degrees)
# winds = ??
# windstoot = maximum gust of wind (in Beaufort scale??)
# temp = measured temperature
# mintemp = measured minimum temperature
# dauwpunt = measured dew point
# zonduur = total minutes of sunshine (in the specific hour)
# straling = the uv radiation
# neerslduur = the duration of rain
# neerslsom = mm of rain
# luchtdruk = airpressure in hPa
# zicht = sight distance in meters
# bewolking = not sure but this is related to the clouds
# rv = relative humidity
# code = ??
# mist = dummy variable whether there is fog 
# regen = dummy variable whether there is rain
# sneeuw = dummy variable whether there is snow
# onweer = dummy variable whether there is thunder
# geen_waarneming = idk what this means 

#Make all the values numeric if possible and make copy of dataframe
cols = dataframe.columns[1:]
dataframe[cols] = dataframe[cols].apply(pd.to_numeric, errors='coerce', axis=1)
df = dataframe.copy()

#Convert date to datetime object and plot the date against the temperature
df['datum'] = pd.to_datetime(df['datum']) + pd.to_timedelta(df['uur'], unit='h')
plt.plot(df['datum'], df['temp']);
```
