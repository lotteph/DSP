# Averaging

```python
resono_data = resono_all.copy()
resono_data
```

```python
#Top 6 biggest parks
locations = ['Rembrandtpark Noord',
 'Rembrandtpark Zuid',
 'Sarphatipark',
 'Vondelpark Oost',
 'Vondelpark West',
 'Oosterpark',
 'Erasmuspark',
 'Westergasfabriek']

# Only select resono data of biggest 6 parks
df_average_visits_parks = resono_data[resono_data['Location'].isin(locations)]
```

```python
df_average_visits_parks.groupby('Location').count()
```

```python
df_average_visits_parks['Visits'] = df_average_visits_parks['Visits'].astype(float, errors = 'raise')
```

```python
df_average_visits_parks.info()
```

```python
df_average_visits = df_average_visits_parks 
df_average_visits['Start'] = pd.to_datetime(df_average_visits['Start'])
df_average_visits = df_average_visits.groupby(['Location', 'Start']).mean().reset_index()
df_average_visits['End'] = df_average_visits['Start'] + dt.timedelta(minutes=14, seconds=59)
```

```python
df_average_visits.loc[(df_average_visits['Location'] == 'Vondelpark Oost')|
                      (df_average_visits['Location'] == 'Vondelpark West')]


#Vondelpark Oost = 8876 rows
#Vondelpark West = 8876 rows
```

```python
# Merge Noord-Zuid & Oost-West into one location
df_average_visits_parks.loc[df_average_visits_parks.Location.str.contains('Rembrandtpark'), 'Location'] = 'Rembrandtpark'
df_average_visits_parks.loc[df_average_visits_parks.Location.str.contains('Vondelpark'), 'Location'] = 'Vondelpark'

df_average_visits_parks_summed = df_average_visits_parks.groupby(['Location', 'Start', 'End']).sum().reset_index()
```

```python
df_average_visits_parks_summed.loc[(df_average_visits_parks_summed['Location'] == 'Vondelpark')]
#Vondelpark = 8876 rows
```

```python
resono = df_average_visits_parks_summed
resono.groupby('Location').count()
```

```python
resono['Start_Dates'] = pd.to_datetime(resono['Start']).dt.date
resono['Start_Time'] = pd.to_datetime(resono['Start']).dt.time

resono['End_Dates'] = pd.to_datetime(resono['End']).dt.date
resono['End_Time'] = pd.to_datetime(resono['End']).dt.time
```
