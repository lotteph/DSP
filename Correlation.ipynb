```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")
```

```python
complete_df = pd.read_csv("Complete dataframe/data_preprocessing.csv")
```

```python
complete_df["Datetime"] = pd.to_datetime(complete_df["Datetime"])
complete_df = complete_df.set_index("Datetime")
complete_df_2021_2022 = complete_df.loc['2021':'2022'].drop(columns=['Year'])
complete_df_2021_2022.head()
```

```python
# Compute the correlation matrix
corr = complete_df_2021_2022.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
