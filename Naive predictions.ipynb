```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import seaborn as sns
```

```python
complete_df = pd.read_csv("Complete dataframe/data_preprocessing.csv")
complete_df["Datetime"] = pd.to_datetime(complete_df["Datetime"])
complete_df = complete_df.set_index("Datetime")
complete_df_2021_2022 = complete_df.loc['2021':'2022'].drop(columns=['Year'])
complete_df_2021_2022.head()
```

```python
# Linear regression
X = np.array(complete_df_2021_2022.loc[:, 'Journeys' : 'stringency_index'])
y = np.array(complete_df_2021_2022.loc[:, "Visits"]).reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor = LinearRegression()

regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

y_test.reshape(-1)

predictions_actual = pd.DataFrame({'Actual': y_test.reshape(-1), 'Predicted': y_pred.reshape(-1)})

rmse = np.sqrt(mean_squared_error(predictions_actual['Actual'], predictions_actual['Predicted']))
mae = mean_absolute_error(predictions_actual['Actual'], predictions_actual['Predicted'])
print("RMSE : % f" %(rmse))
print("MAE : % f" %(mae))
```
